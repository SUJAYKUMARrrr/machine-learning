```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = [
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]

y = [0,0,1,1] # labels 0 = apple, 1 = orange

model = DecisionTreeClassifier()
model.fit(x,y)

size = float (input("enter the fruit size in cm: "))
shade = float (input("enter the fruit shade (1-10): "))

prediction = model.predict([[size,shade]])
#model.predict([[280,8.3]]) [0] > 1 [1]


if prediction == 0:
    print("The fruit is an apple")
else:
    print("The fruit is an orange")
```
