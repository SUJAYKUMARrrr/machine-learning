```python
"""
linear regression
1- find a pattern in old data
2- use the pattern to predict future data
3- straight line equation: y = mx + b
"""
from sklearn.linear_model import LinearRegression

X = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]

model = LinearRegression()
model.fit(X,y)
hours = float(input("Enter the number of hours studied: "))
predicted_score = model.predict([[hours]])

print(f"Predicted score for studying {hours} hours: {predicted_score[0]}")

```
