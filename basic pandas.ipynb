```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age' : [25, 30, None, 40],
    'salary': [50000, 60000, 55000, None]}
df = pd.DataFrame(data)
```

```python
data
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
df_dropped = df.dropna()
print(df_dropped)
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['salary'].fillna(df['salary'].mean(), inplace=True)
print(df)
```
